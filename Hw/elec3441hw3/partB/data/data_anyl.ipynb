```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

# 读取CSV文件
data = pd.read_csv('result.csv')

# 提取数据列
length = data['length']
n_cubed = np.power(data[' N^3'], 1/3)
ijk_time = np.power(data[' ijk'], 1/3) / n_cubed
kij_time = np.power(data[' kij'], 1/3) / n_cubed
jki_time = np.power(data[' jki'], 1/3) / n_cubed

# 绘制折线图
plt.plot(length, ijk_time, label='ijk')
plt.plot(length, kij_time, label='kij')
plt.plot(length, jki_time, label='jki')

# 设置图表标题和坐标轴标签
plt.title('Time Taken for Matrix Multiplication')
plt.xlabel('Length')
plt.ylabel('Time Taken')

# 设置横轴刻度间隔为10
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(50))
plt.xlim(20, max(length))  # 设置横轴范围从20开始
plt.xticks(rotation=270)

# 添加图例
plt.legend()

# 显示图表
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

# 读取CSV文件
data = pd.read_csv('result-0-40-10.csv')

# 提取数据列
length = data['length']
n_cubed = np.power(data[' N^3'], 1/3)
ijk_time = np.power(data[' ijk'], 1/3) / n_cubed
kij_time = np.power(data[' kij'], 1/3) / n_cubed
jki_time = np.power(data[' jki'], 1/3) / n_cubed

# 绘制折线图
plt.plot(length, ijk_time, label='ijk')
plt.plot(length, kij_time, label='kij')
plt.plot(length, jki_time, label='jki')

# 设置图表标题和坐标轴标签
plt.title('Normalized Time Taken for Matrix Multiplication')
plt.xlabel('Length')
plt.xticks(rotation=270)
plt.ylabel('Normalized Time Taken')

# 设置横轴刻度间隔为20
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(40))

# 添加图例
plt.legend()

# 显示图表
plt.show()
```
